```python
x=df.iloc[:,5:33]
```

```python
from sklearn.decomposition import PCA
```

```python
pca=PCA(n_components=2,random_state=42)
```

```python
x_pca=pca.fit_transform(x)
```

```python
x_pca
```

```python
pca.explained_variance_ratio_.cumsum()[1]
```

## Model Training

```python
from sklearn.cluster import KMeans
```

```python
distortions=[]
cluster_range=range(1,6)
for i in cluster_range:
    model=KMeans(n_clusters=i,init='k-means++', random_state=42)
    model.fit(x_pca)
    distortions.append(model.inertia_)
```

```python
plt.plot(range(1, 6), distortions,marker='o')
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('Distortions')
plt.show()
```

```python
model=KMeans(n_clusters=3,init='k-means++', random_state=42)
model.fit(x_pca)
y=model.predict(x_pca)
```

```python
plt.scatter(x_pca[y==0,0],x_pca[y==0,1],s=100,color='red',label='Cluster-1')
plt.scatter(x_pca[y==1,0],x_pca[y==1,1],s=100,color='green',label='Cluster-2')
plt.scatter(x_pca[y==2,0],x_pca[y==2,1],s=100,color='yellow',label='Cluster-3')
plt.scatter(model.cluster_centers_[:, 0], model.cluster_centers_[:, 1], s = 300, c = 'blue', label = 'Centroids')
plt.title('Clusters of students')
plt.xlabel('PCA 1')
plt.ylabel('PCA 2')
plt.legend()
plt.show()
```

```python
from collections import Counter
Counter(y)
```

```python
import scipy.cluster.hierarchy as hier
dendrogram = hier.dendrogram(hier.linkage(x_pca, method = 'ward'))
plt.title('Dendrogram')
plt.xlabel('questions')
plt.ylabel('Euclidean distances')
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
hc = AgglomerativeClustering(n_clusters = 2, affinity = 'euclidean', linkage = 'ward')
y_hc = hc.fit_predict(x_pca)
X = x_pca
# Visualising the clusters
plt.scatter(X[y_hc == 0, 0], X[y_hc == 0, 1], s = 100, c = 'yellow', label = 'Cluster 1')
plt.scatter(X[y_hc == 1, 0], X[y_hc == 1, 1], s = 100, c = 'red', label = 'Cluster 2')
plt.title('Clusters of STUDENTS')
plt.xlabel('PCA 1')
plt.ylabel('PCA 2')
plt.legend()
plt.show()
```
